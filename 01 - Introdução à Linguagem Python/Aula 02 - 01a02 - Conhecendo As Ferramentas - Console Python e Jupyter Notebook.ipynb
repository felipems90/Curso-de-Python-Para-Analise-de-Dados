# Apresentação do Google Colaboratory

### Ele também é gratuito 
O Colab é gratuito para usar, mas os recursos de processamento não são garantidos e ilimitados, e os limites de uso às vezes variam. Isso é necessário para que o Colab possa fornecer recursos gratuitamente.  Os usuários interessados ​​em um acesso mais confiável a melhores recursos podem se interessar pelo Colab Pro .


###  A diferença entre Jupyter e Colab
O Google Colaboratory é muito similar ao Jupyter notebook, com excessão das funções específicas para seu uso em nuvem.

Jupyter é o projeto de código aberto no qual o Colab se baseia. O Colab permite que você use e compartilhe notebooks Jupyter com outras pessoas sem precisar baixar, instalar ou executar nada.

### Armazenamento e compartilhamento do notebooks do Colab
Os notebooks Colab são armazenados no Google Drive ou podem ser carregados do GitHub . 

Os cadernos do Colab podem ser compartilhados da mesma forma que você faria com o Google Docs ou Planilhas. Basta clicar no botão Compartilhar no canto superior direito de qualquer bloco de notas Colab ou seguir estas instruções de compartilhamento de arquivos do Google Drive.
