# Manipulação de arquivos Python 
O Python tem várias funções para criar, ler, atualizar e excluir arquivos.
A a função open() é a principal para trabalhar com arquivos em Python.

Ela recebe dois parâmetros; nome do arquivo e modo .

### Modos de acesso:

- r"- Ler - (padrão) Abre um arquivo para leitura, gera erro se o arquivo não existe;
- "a" - Adicionar - abre um arquivo para adicionar mais conteúdom, mas cria o arquivo se ele não existir;
- "w" - Gravar - abre um arquivo para gravação, cria o arquivo se ele não existir
- "x" - Criar - Cria o arquivo especificado, retorna um erro se o arquivo existir

### Pode se especificar se o arquivo deve ser tratado como modo binário ou  texto

- "t"- Texto - (padrão);

- "b" - Binário;


```python
conteudo =  os.listdir(diratual)
print(conteudo)
```

# Fechar arquivos após a operação
Sempre feche o arquivo quando terminar de usá-lo, pois devido ao armazenamento em buffer, as alterações feitas em um arquivo podem não aparecer até que você feche o arquivo.

```python
f = open("tabela6579.csv", "r", encoding="utf8")
print(f.readline())
f.close()
```

## Use a cláusula with para que o arquivo seja fechado automaticamente após as operações

```python
with open("tabela6579.csv", "r", encoding="utf8") as f:
  print(f.readline())


print(f.readline())
```

## Gravação de arquivo Python
Gravar em um arquivo existente
Deve-se adicionar um parâmetro à função open():
- "a" - Adicionar - irá anexar ao final do arquivo
- "w" - Gravação - sobrescreverá qualquer conteúdo existente


```python

texto = """Linha 01
Linha 02
Linha 03"""

with open("texto.txt", "w") as f:
  f.write(texto)


with open("texto.txt","r") as f: 
  print(f.read())

```

## Sobrescrevendo o Arquivo

```python
with open("texto.txt","r") as f: 
  print(f.read())
  
texto = """Linha 01
Linha 02
Linha 03"""

with open("texto.txt", "a") as f:
  f.write(texto)


with open("texto.txt","r") as f: 
  print(f.read())


texto = """
Linha 04
Linha 05
Linha 06
"""
with open("texto.txt", "w") as f:
  f.write(texto)


with open("texto.txt","r") as f: 
  print(f.read())
```

## Adicionando conteúdo no final do arquivo

```python
texto = """Linha 01
Linha 02
Linha 03"""

with open("texto.txt", "w") as f:
  f.write(texto)

texto = """
Linha 04
Linha 05
Linha 06"""
with open("texto.txt", "a") as f:
  f.write(texto)


with open("texto.txt","r") as f: 
  print(f.read())
```

```python
with open("texto.txt","r") as f: 
  linhas = f.readlines()
  novoconteudo = [l for l in linhas if int(l[-2:]) % 2 ==0]
  print(novoconteudo)

with open("texto.txt","w") as f:
  f.writelines(novoconteudo)

with open("texto.txt","r") as f: 
  print(f.read())
```

## Verifique se o arquivo existe:


```python
import os
if os.path.exists("texto.txt"):
  os.remove("texto.txt")
else:
  print("O arquivo não existe")
```
