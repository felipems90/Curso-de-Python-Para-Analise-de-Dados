## Utilizando um Mapa de Cores

O Matplotlib possui vários mapas de cores (lista de cores), onde cada cor tem um valor que varia de 0 a 100.

Para usar o mapa de cores utilize parâmetro  cmapcom com o nome do mapa e no parâmetro c passe o array de valores para cada variando de 0 a 100


Para cada mapa de cores existe seu correspondente invertido.

```python
import matplotlib.pyplot as plt
import numpy as np

xcasos = np.log(totaisestados_arr[:,1].astype('i'))
yobitos = np.log(totaisestados_arr[:,2].astype('i'))
min = np.min(yobitos)
max = np.max(yobitos)
rng = max-min
cores = np.array([int((x - min) /rng*100)  for x in yobitos])
print(cores)

plt.scatter(xcasos, yobitos, c=cores, cmap='viridis')
plt.show()
```

```python
plt.scatter(xcasos, yobitos, c=cores, cmap='viridis_r')
plt.show()
```

### Incluindo a barra de cores com a função plt.colorbar()


```python
plt.scatter(xcasos, yobitos, c=cores, cmap='viridis_r')

plt.colorbar()

plt.show()
```
