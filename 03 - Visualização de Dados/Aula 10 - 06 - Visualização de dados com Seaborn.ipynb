# Gráficos com Seabor

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white" )
plt.rcParams["figure.figsize"] = [6, 4]

x= total_brasil[:,0]
y= total_brasil[:,1].astype('i')/1000

fig = sns.lineplot(x=x, y=y, marker='o', linestyle='dashed')

plt.title('Evolução dos Casos de COVID-19 no Brasil (Em milhares)')
plt.xlabel('Mês')
plt.ylabel('Quantidade')
plt.ylim(bottom=0)
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white" )

xy = sorted(zip(totaisregioes_arr[:,0], totaisregioes_arr[:,1].astype('i')/1000), key=lambda pair: pair[1], reverse=True)
x= [x for x , _ in xy]
y= [y for _ , y in xy]
print(xy)
fig , (ax1, ax2) = plt.subplots(2, 1)

sns.barplot(x=x, y=y,palette="rocket", ax=ax1)
ax1.set_title('Evolução dos Casos de COVID-19 no Brasil (Em milhares)')
#ax1.set_xlabel('Região')
ax1.set_ylabel('Quantidade')
ax1.set_ylim(bottom=0)

sns.barplot(x=y, y=x,palette="rocket", ax=ax2 , orient='horizontal', dodge=False)
ax2.set_xlabel('Quantidade')
ax2.set_ylabel('Região')

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
plt.rc('figure', figsize = (8.27, 11.7))
xy = sorted(zip(totaisregioes_arr[:,0], totaisregioes_arr[:,1].astype('i')/1000), key=lambda pair: pair[1], reverse=True)
labels= [x for x , _ in xy]
sizes= [y for _ , y in xy]
print(xy)

fig , (ax1, ax2) = plt.subplots(2, 1)

theme = plt.get_cmap('copper')
cores = [theme(1 * i / len(sizes)) for i in range(1,len(sizes)+1)]
ax1.set_title('Proporção de Casos de COVID-19 no Brasil por Região')
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', 
        colors =cores,  textprops={'color':"w"})
ax1.axis('equal')
ax1.legend(labels,
          loc="lower left")


ax2.pie(sizes, labels=labels, autopct='%1.1f%%', 
        colors =cores,  textprops={'color':"w"},
        wedgeprops=dict(width=0.6))
ax2.axis('equal')


plt.show()
```
