### Parâmetros da área de plotagem dos gráficos

- figsize: (float, float), opcional -  Tamanho da figura. Se não for fornecido, o padrão é rcParams ["figure.figsize"] = [6.4, 4.8].

````python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"] = [6, 4]
xpoints = np.array([10, 100])
ypoints = np.array([50, 250])

plt.plot(xpoints, ypoints)
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"] = [2, 2]
xpoints = np.array([10, 100])
ypoints = np.array([50, 250])

plt.plot(xpoints, ypoints)
plt.show()
````

#### dpi: inteiro, opcional -  Resolução da figura. Se não for fornecido, o padrão é rcParams ["figure.dpi"] = 100.

````python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.dpi"] = 300
xpoints = np.array([10, 100])
ypoints = np.array([50, 250])

plt.plot(xpoints, ypoints)
plt.show()
````

#### facecolor: especificação de cor - Cor de fundo. Se não for fornecido, o padrão é rcParams ["figure.facecolor"] = 'w'.

Alias das cores
- 'b'	blue
- 'g'	green
- 'r'	red
- 'c'	cyan
- 'm'	magenta
- 'y'	yellow
- 'k'	black
- 'w'	white

- Valor hexadecimal da cor '#eeefff'
- Ou qualquer nome de cores HTML 
https://www.homehost.com.br/blog/tutoriais/tabela-de-cores-html/#topico1

````python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = [2, 2]
plt.rcParams["figure.dpi"] = 100
plt.rcParams["figure.facecolor"] = 'MediumSeaGreen'
xpoints = np.array([10, 100])
ypoints = np.array([50, 250])

plt.plot(xpoints, ypoints)
plt.show()
````

## Desenhar vários gráficos juntos
A função subplots() é usada para desenhar vários gráficos em uma figura

A função subplots() leva três argumentos: primeiro quantidade de linhas, segundo quantidade de colunas, terceiro  o índice do gráfico atual.



### Para desenhar dois gráficos

````python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = [10, 6]
xmeses = total_brasil[:,0]
ycasos = total_brasil[:,1].astype('i')
yobitos = total_brasil[:,2].astype('i')

plt.subplot(1, 2, 1)
plt.plot(xmeses,ycasos)


plt.subplot(1, 2, 2)
plt.plot(xmeses,yobitos)

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

xmeses = total_brasil[:,0]
ycasos = total_brasil[:,1].astype('i')
yobitos = total_brasil[:,2].astype('i')

plt.subplot(2, 1, 1)
plt.plot(xmeses,ycasos)


plt.subplot(2, 1, 2)
plt.plot(xmeses,yobitos)

plt.show()
````

## Podem ser desenhados quantos gráficos quiser em uma figura


````python
import matplotlib.pyplot as plt
import numpy as np
xmeses = total_brasil[:,0]
ycasos = total_brasil[:,1].astype('i')
yobitos = total_brasil[:,2].astype('i')

plt.subplot(2, 3, 1)
plt.plot(xmeses,ycasos, 'o-r')

plt.subplot(2, 3, 2)
plt.plot(xmeses,ycasos, '*:b')

plt.subplot(2, 3, 3)
plt.plot(xmeses,ycasos, 's--m')


plt.subplot(2, 3, 4)
plt.plot(xmeses,yobitos, 'o-r')

plt.subplot(2, 3, 5)
plt.plot(xmeses,yobitos, '*:b')

plt.subplot(2, 3, 6)
plt.plot(xmeses,yobitos, 's--m')

plt.show()
````

### Adicionar um título a cada gráfico com a função title()

````python
import matplotlib.pyplot as plt
import numpy as np

xmeses = total_brasil[:,0]
ycasos = total_brasil[:,1].astype('i')
yobitos = total_brasil[:,2].astype('i')

plt.subplot(2, 1, 1)
plt.plot(xmeses,ycasos)
plt.title("Quantidade de Casos")

plt.subplot(2, 1, 2)
plt.plot(xmeses,yobitos)
plt.title("Quantidade de Óbitos")
plt.show()
````

### Adicionar um título a toda a figura com a função suptitle()


````python
import matplotlib.pyplot as plt
import numpy as np

xmeses = total_brasil[:,0]
ycasos = total_brasil[:,1].astype('i')
yobitos = total_brasil[:,2].astype('i')

plt.subplot(2, 1, 1)
plt.plot(xmeses,ycasos)
plt.title("Quantidade de Casos")

plt.subplot(2, 1, 2)
plt.plot(xmeses,yobitos)
plt.title("Quantidade de Óbitos")


plt.suptitle("Dados da Covid-19 no Brasil")
plt.show()
````

### Ajustando a distancia entres os gráficos com a função plt.subplots_adjust()

subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)


````python
import matplotlib.pyplot as plt
import numpy as np

xmeses = total_brasil[:,0]
ycasos = total_brasil[:,1].astype('i')
yobitos = total_brasil[:,2].astype('i')

plt.subplot(2, 2, 1)
plt.plot(xmeses,ycasos)
plt.title("Quantidade de Casos")

plt.subplot(2, 2, 2)
plt.plot(xmeses,yobitos)
plt.title("Quantidade de Óbitos")

plt.subplot(2, 2, 3)
plt.plot(xmeses,ycasos)
plt.title("Quantidade de Casos")

plt.subplot(2, 2, 4)
plt.plot(xmeses,yobitos)
plt.title("Quantidade de Óbitos")

plt.subplots_adjust( wspace = 0.2, hspace = 0.4)
plt.suptitle("Dados da Covid-19 no Brasil")
plt.show()
````
